Now that we are authenticated. We are going to look at the bikers trip data from Austin. We have a project called data-viz-bishal in my BigQuery interface.

1. What was the second most popular starting station?

2. How many trips lasted over an hour and were a round trip (started and stopped at the same station)?


```python
%%bigquery --project data-viz-bishal
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
LIMIT 10
```

The above dataset is the data from bikershare trips of start time and end time with start station and end station.

```python
%%bigquery --project data-viz-bishal
SELECT MAX(start_station_name) as second_most_popular
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE start_station_name != (SELECT MAX(start_station_name) as most_popular_start_station FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`)
```

The most second popular starting station is Zilker Park at Barton Springs and William Barton.

```python
%%bigquery --project data-viz-bishal
SELECT  COUNTIF(duration_minutes>=60) as over_an_hour_duration ,COUNT( start_station_name) as number_of_trips
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE start_station_name = end_station_name
```

228339 trips lasted over an hour and were a round trip (started and stopped at the same station).

```python
%%bigquery --project data-viz-bishal
SELECT  start_station_name as station_name,  COUNTIF(duration_minutes>=60) as over_an_hour_duration, COUNT( start_station_name) as number_of_trips
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE start_station_name = end_station_name
GROUP BY start_station_name
```

The above dataset is the calculation for the ways to get trips lasted over an hour and for the round trip.
