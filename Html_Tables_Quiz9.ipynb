Navigate to the wikipedia page on Marvel Cinematic Universe Films. Gather the table on the films in the Infinity series (Hint: class is 'wikitable plainrowheaders'). Fix any issues with the column names. Remove rows that are not movies.

Loading necessary libraries

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

As instructed using the wikipedia link https://en.wikipedia.org/wiki/List_of_Marvel_Cinematic_Universe_films to gather tables.

```python
r = requests.get('https://en.wikipedia.org/wiki/List_of_Marvel_Cinematic_Universe_films')
html_parse = r.text
html_soup = BeautifulSoup(html_parse,'lxml')
```

```python
len(html_soup.find_all('table'))
```

There are 7 tables in a list.
Now,getting the table from the webpage.

```python
marvelTable = html_soup.find_all('table', class_ = 'wikitable plainrowheaders')
dataFrame = pd.read_html(str(marvelTable))[0]
dataFrame
```

Removing unnecessary rows over the heading using droplevel method in pandas.

```python
dataFrame.columns = dataFrame.columns.droplevel(1)
dataFrame
```

Looking at the table we can see row 6 and 13 are not movies so we have to remove those two rows using the same drop method from pandas. As we do that it messes up the order of the rows so we have to fix that too.

```python
dataFrame = dataFrame.drop(dataFrame.index[[6,13]])
dataFrame
```

```python
dataFrame = dataFrame.reset_index(drop=True, inplace=False)
dataFrame
```

Now, we will remove the column Unnamed: 5_level_0 from the index as it does not reveal kind of information.

```python
del dataFrame['Unnamed: 5_level_0']
dataFrame
```

So, we have removed a column and 2 rows from the wikipedia link: 
https://en.wikipedia.org/wiki/List_of_Marvel_Cinematic_Universe_films
