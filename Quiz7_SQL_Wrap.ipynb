How many trips have excatly thirty minutes duration which starts and ends at the same station and subscriber_type is Walk_up ?

Lets print first five data to see what type of categories are in the dataset.

```python
%%bigquery --project data-viz-bishal
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
LIMIT 5
```

From this dataset we will be using subsciber_type, start_station_name,end_station_name and duration_minutes in this project.

To begin with, lets find out how many trips have duration of excatly 30 mins.

```python
%%bigquery --project data-viz-bishal
SELECT  COUNTIF(duration_minutes=30) as Thirty_Minutes_duration
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
```

So, 12236 trips have 30 minutes duration.

Now to complicate this problem and trim this data even more, We are going to find out how many trips have thirty minutes. duration which starts and ends at the same station.

```python
%%bigquery --project data-viz-bishal
SELECT COUNT(*) as Thirty_MinuteS_Trips
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE start_station_name = end_station_name AND duration_minutes =30
```

So this has excluded many trips. only 2476 trips starts and ends at same station which have 30 minutes duration.

Lets trim this even more and bring subscirber type to this equation. Lets find out how many trips have thirty minutes duration which starts and ends at the same station and subscriber_type is Walk_up

```python
%%bigquery --project data-viz-bishal
SELECT COUNT(*) as Thirty_MinuteS_Trips_AND_WALK_UP
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE duration_minutes =30 AND subscriber_type ="Walk Up" AND start_station_name = end_station_name
```

From this output, We found out that less than 1000 trips fall into the category we are looking for.
