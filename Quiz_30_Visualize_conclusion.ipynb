```python
import matplotlib
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv')

df.groupby(['neighbourhood_group','room_type']).price.agg('max').plot(kind = 'barh',
                                                 title = 'Max price', 
                                                 ylabel= 'Price in dollar')
```

The above dataset is NYC air b&b data to create a multiple bar graph with 'neighbourhood_group' and 'room_type' by looking at the average 'price'. Another important graph is histogram can asked in Exam 3.
