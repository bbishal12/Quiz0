Does converting SepalLength to integer increase or decrease the mean?

Does the direction of the shift remain the same if you groupby Class?

Gather the mean, median, count and standard deviation of all columns when grouped by Class.

```python
import pandas as pd
import statistics

iris_df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/iris.csv')
iris_df
```

```python
from numpy.ma.core import mean
mean(iris_df.SepalWidth)
```

```python
new_df= iris_df.SepalWidth.astype('int') 
new_df
```

```python
mean(new_df)
```

```python
rnew_df=round(iris_df.SepalWidth).astype('int')
rnew_df
```

```python
mean(rnew_df)
```

Now, applying the shift function.

```python
df_test=iris_df.shift()
df_test
```

```python
col= iris_df.groupby('Class')
col.shift()
```

Now, lets find out mean, median, count and std of the class.

```python
iris_df.groupby('Class').SepalWidth.agg(['mean','median','count', 'std'])
```

The above data  is the mean, median, count and std of all columns by class.
